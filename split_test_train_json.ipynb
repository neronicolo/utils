```python
import json
from pathlib import Path
```

```python
source_dir = Path('/home/nikola/projects/pxo/data/prepared/nerfstudio/living_room/')
path_json = source_dir/'transforms.json'

# Open the JSON file
with open(path_json) as file:
    # Load the JSON data into a dictionary
    json_data = json.load(file)

print(json_data)
```

```python
frames = json_data["frames"]
frames_train = []
frames_test = []

for i,frame in enumerate(frames):

    if i % 8 != 0:
        frames_train.append(frame)
    else:
        frames_test.append(frame)

print(f'frames_len:{len(frames)}')
print(f'frames_train_len:{len(frames_train)} \nframes_test_len:{len(frames_test)}')        
```

```python
# Copy dictionary
json_data_train = json_data.copy()
json_data_test = json_data.copy()

# Modify the values in the dictionary
json_data_train['frames'] = frames_train
json_data_test['frames'] = frames_test

print(f'json_data_train:{json_data_train["frames"]} \njson_data_train:{json_data_test["frames"]}')
```

```python
# Dictionary containing data for each JSON file
json_data_files = {
    'transforms_train.json':json_data_train, 
    'transforms_test.json':json_data_test
}
```

```python
# Save each JSON file
for file_name, data in json_data_files.items():
    with open(source_dir/file_name, 'w', encoding="utf-8") as file:
        json.dump(data, file, indent=4)
```
