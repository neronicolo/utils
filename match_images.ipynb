```python
import json
import shutil
from pathlib import Path
```

```python
source_dir = Path('/home/nikola/projects/pxo/data/outputs/fort_langley_train_nerfstudio/')
path_json = source_dir/'transforms.json'
images_dir = source_dir/'images/'
target_dir = Path('/home/nikola/projects/pxo/data/outputs/fort_langley_train_llff/images/')

# Open the JSON file
with open(path_json) as file:
    # Load the JSON data into a dictionary
    json_data = json.load(file)

print(json_data)
```

```python
frames = json_data['frames']
file_names = []

for item in frames:
    file_name = images_dir/Path(item['file_path']).name
    file_names.append(file_name)

file_names.sort()

target_dir.mkdir(parents=True, exist_ok=True)

for file_name in file_names:
    source_file = Path(file_name)
    if source_file.is_file():  # check if file exists
        shutil.copy(source_file, target_dir)  # copy the file to the target directory
    else:
        print(f"The file {file_name} does not exist")
```
